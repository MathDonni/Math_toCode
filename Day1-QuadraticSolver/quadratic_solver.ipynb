# Quadratic Equation Solver  
This notebook demonstrates how to solve quadratic equations using Python.  

## Math Recap  
A quadratic equation is of the form:  
\[
`ax^2 + bx + c = 0`
\]  `
The roots are given by:  
\[
`x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
\] `

```python
# Python Implementation
import math

def quadraticSolver(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return "No real roots"
    else:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return root1, root2

# Example
a, b, c = 1, -5, 6
roots = quadraticSolver(a, b, c)
print(f"Roots: {roots}")
```

## Visualization  
Let’s plot the quadratic equation \( y = x^2 - 5x + 6 \):  

```python
# Plotting
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 6, 100)
y = a*x**2 + b*x + c

plt.plot(x, y, label=f"y = {a}x² + {b}x + {c}")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.scatter(roots, [0, 0], color="red", label="Roots")
plt.legend()
plt.title("Quadratic Equation Graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
